Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.replace('NaN', np.nan, inplace=True)
```

```python
display(df.isnull().sum()/len(df)*100)
```

Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"

```python
a = df.loc[:, "N":"Язык.программирования"]
b = df.loc[:, "Специализация":"cls"].dropna(axis=1)
```

```python
dat = pd.concat([a,b], axis=1)
np.round(dat.isnull().sum()/len(dat)*100, 0)
```

Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

Визначте новий розмір таблиці за допомогою методу shape

```python
dat = dat.dropna()
dat.shape
```

Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data = dat[dat['Язык.программирования'] == 'Python']
python_data.shape
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
group_data = python_data.groupby(['Должность'])
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary_data = group_data.agg({"Зарплата.в.месяц": ['min', 'max']})
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(x):
  return np.mean(x)

salary_data['avg'] = salary_data['Зарплата.в.месяц'].apply(fill_avg_salary, axis = 1)
display(salary_data)
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
np.round(salary_data['avg'].describe(),2)
```

Збережіть отриману таблицю в CSV файл

```python
salary_data.to_csv('Salary_data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
c_data = dat[dat['Язык.программирования'] == '1С']
job_data = c_data['Должность'].value_counts()
plt.title('Розподіл 1C програмістів за посадами')
plt.pie(job_data,
        labels=job_data.index,
        autopct="%.2f%%",
        explode = [0.06, 0.06, 0.06, 0.06],
        )

plt.show()
```

```python
import plotly.express as px
```

```python
city_data = dat.groupby('Город')
salary_data = city_data.agg({"Зарплата.в.месяц": 'mean'})
fig = px.line(salary_data, x=salary_data.index, y="Зарплата.в.месяц", title = "Средня зарплата програмістів по містах")
fig.show()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=dat.sort_values(by=['Должность', 'Город'], ascending=True), x='Город', y='Зарплата.в.месяц',
              hue='Должность'
              )
plt.title('Зарплати програмістів по містам, залежно від посади')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
plt.show()
```

```python
study_data = dat.groupby('Образование')
study_data = study_data.agg({"Зарплата.в.месяц": 'mean'})
sns.barplot(data=study_data, y='Зарплата.в.месяц', x=study_data.index)
plt.title('Середня зарплата програмістів, залежно від освіти')
plt.xticks(rotation=90)
plt.show()
```
