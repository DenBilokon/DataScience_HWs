```python
import numpy as np
import pandas as pd
```

Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
data_url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
match_text = "Коефіцієнт народжуваності у регіонах України у 1950—2014 роках"
data = pd.read_html(data_url, match=match_text)
df = data[0]
display(df)
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
df.replace('NaN', np.nan, inplace=True)
display(df)
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum()
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['1950'] = df['1950'].fillna(df['1950'].mean()).astype('int64')
df['1960'] = df['1960'].fillna(df['1960'].mean()).astype('int64')
df['1970'] = df['1970'].fillna(df['1970'].mean()).astype('int64')
df['2014'] = df['2014'].fillna(df['2014'].mean()).astype('int64')
df.dtypes
display(df)
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop(len(df)-1)
display(df)
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df[df['2014']>df['2014'].mean()]
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

У якому регіоні була найвища народжуваність у 2014 році?

```python
df.loc[df['2014'] == df['2014'].max()].iloc[:, 0]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(
    df['регіон'],
    df['2014'],
    color = "indigo"
    )
plt.xlabel("Регіон", fontsize=10, color="indigo")
plt.ylabel("Рівень народжуваності", fontsize=10, color="indigo")
plt.title("Діаграма народжуваності по регіонах у 2014 році", fontsize=15, color="indigo")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Додаткова візуалізація:

```python
df['avg'] = df.mean(axis=1)
```

```python
import plotly.express as px
```

```python
fig = px.line(df, x='регіон', y="avg", title = "Середня народжуваність по регіонам")
fig.show()
```

```python
import matplotlib.pyplot as plt
```

```python
df.set_index('регіон').T.plot(kind='line', marker='o', figsize=(15, 6))
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.title('Графік народжуваності в Україні в розрізі кожного регіону з 1950 по 2014 роки')
plt.show()
```
